## CONCENTRIC CIRCLES - BASIC

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_concentric_circles(labels, radii, figsize=(10, 10)):
    """Plot concentric circles with given labels and radii."""
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_aspect('equal', 'box')
    center = (0, 0)
    
    for label, radius in zip(labels, radii):
        circle = patches.Circle(center, radius, fill=False)
        ax.add_patch(circle)
        
        # Place the label in the middle of the current circle's segment
        ax.text(center[0], center[1] + radius - (0 if len(radii) == 1 else (radii[1]-radii[0])/2), 
                label, ha='center', va='center')
    
    ax.set_xlim(-max(radii)-1, max(radii)+1)
    ax.set_ylim(-max(radii)-1, max(radii)+1)
    ax.axis('off')
    plt.show()

# Example usage
labels = ['Inner', 'Middle', 'Outer']
radii = [5, 10, 15]
plot_concentric_circles(labels, radii, figsize=(12, 12))  # Adjust the figsize values as needed
```

## CONCENTRIC CIRCLES - DYNAMIC

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches


def plot_concentric_circles(labels, circle_colors, font_colors, font_size=12, figsize=(10, 10)):
    """Plot concentric circles with given labels."""
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_aspect('equal', 'box')
    center = (0, 0)
    radii = [5 + i*5 for i in range(len(labels))]  # Increasing radius for each circle by 5 units
    
    # First, draw and fill the circles from largest to smallest
    for radius, circle_color in sorted(zip(radii, circle_colors), reverse=True):
        circle = patches.Circle(center, radius, fill=True, facecolor=circle_color, edgecolor='black', linewidth=1.5)
        ax.add_patch(circle)

    # Then, add the labels
    for label, radius, font_color in zip(labels, radii, font_colors):
        ax.text(center[0], center[1] + radius - (0 if len(radii) == 1 else (radii[1]-radii[0])/2), 
                label, ha='center', va='center', color=font_color, fontsize=font_size)
    
    ax.set_xlim(-max(radii)-1, max(radii)+1)
    ax.set_ylim(-max(radii)-1, max(radii)+1)
    ax.axis('off')
    plt.show()

# User input
num_circles = int(input("Enter number of circles: "))
labels = [input(f"Name for circle {i+1}: ") for i in range(num_circles)]

# Example color lists (You can customize these as you like)
circle_colors = ['#009933', '#00cc99', '#00bfff', '#0080ff']
font_colors = ['black', 'white', 'black', 'white', 'black', 'white', 'black']

# Cycling through colors based on number of circles (modulus used to loop back to the start if we run out of colors)
circle_colors = [circle_colors[i % len(circle_colors)] for i in range(num_circles)]
font_colors = [font_colors[i % len(font_colors)] for i in range(num_circles)]

plot_concentric_circles(labels, circle_colors, font_colors, font_size=15, figsize=(12, 12))
```
